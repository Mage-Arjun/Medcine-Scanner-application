```python
df = df.merge(
    marketers[["medicine_url", "marketer"]],
    left_on="url",
    right_on="medicine_url",
    how="left"
)

df.drop(columns=["medicine_url"], inplace=True)
```

```python
df.head()
```

```python
# Drop old empty column
df.drop(columns=["marketer_x"], inplace=True)

# Rename merged column
df.rename(columns={"marketer_y": "marketer"}, inplace=True)
```

```python
df["marketer"].isna().mean()
```

```python
def clean_text(x):
    if pd.isna(x):
        return x
    return " ".join(x.split())

df["uses"] = df["uses"].apply(clean_text)
df["side_effects"] = df["side_effects"].apply(clean_text)
```

```python
import re

def parse_composition(comp):
    if pd.isna(comp):
        return []
    return re.findall(r'([A-Za-z]+)\s*\(([\d\.]+mg)\)', comp)

df["composition_parsed"] = df["composition"].apply(parse_composition)
```

```python
df.drop_duplicates(subset=["url"], inplace=True)
```

```python
df.isna().mean().sort_values(ascending=False)
```

```python
df["side_effects"] = df["side_effects"].fillna("Not available")
```

```python
df.to_csv("../data/final/medicine_final_clean.csv", index=False)
```
